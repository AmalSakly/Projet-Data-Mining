```python
#mettre tout en minuscule
tweets = [tweet.lower() for tweet in rgtweets]
tweets
```

**Retrait des ponctuations :**

```python
import string
ponctuations = list(string.punctuation)
print(ponctuations)
roland = ["".join([w for w in list(tweet) if not w in ponctuations]) for tweet in tweets]
roland
```

**Retrait des chiffres et de rt :**

```python
num=[]
for line in roland:
      num.append(line.replace('0',' ').replace('1',' ').replace('2',' ').replace('3',' ').replace('4',' ').replace('5',' ').replace('6',' ').replace('7',' ').replace('8',' ').replace('9',' '))
num 
```

```python
rt=[]
for line in num:
      rt.append(line.replace('rt',' '))
rt 
```

**Retrait des emojis :**

```python
import re


sansemoj=[]
for x in rt:
  def deEmojify(x):
      emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
      
      return emoj.sub(r'',x)

  sansemoj.append(deEmojify(x))
sansemoj
```

**Tokenization :**

```python
from nltk.tokenize import sent_tokenize
sentences = [sent_tokenize(tweet) for tweet in sansemoj]
sentences
```

```python
from nltk.tokenize import word_tokenize
tweetword = [word_tokenize(tweet) for tweet in sansemoj]
tweetword
```

**Retrait des stopwords :**

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
print(stop_words)
```

```python
nostop=[]
for w in tweetword:
  filtwords=[]
  for x in w:
    if x not in stop_words:
        filtwords.append(x)
  nostop.append(filtwords)

nostop
```

```python
processed=[]
for w in nostop:
  ht=[]
  for x in w:
    if not x.startswith('htt'):
        ht.append(x)
  processed.append(ht)

processed
```

```python
import random

random.shuffle(processed)

train_data = processed[:80]
test_data = processed[:20]
```

```python
filtwords=[]
for w in tweetword:
  for x in w:
    if x not in stop_words:
        filtwords.append(x)
filtwords
```
